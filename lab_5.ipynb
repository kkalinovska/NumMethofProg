```python
import numpy as np 
import matplotlib.pyplot as plt 

x_min, x_max = -1, 3 
y_min, y_max = -1, 2 
step = 0.01 

x, y = np.meshgrid(np.arange(x_min, x_max, step), 
                   np.arange(y_min, y_max, step)) 

eq1 = np.cos(x)+ y - 1.5 
eq2 = 2*x - np.sin(y - 0.5) -1 

fig, ax = plt.subplots(figsize=(10, 10)) 
ax.contour(x, y, eq1, levels=[0], colors='red') 
ax.contour(x, y, eq2, levels=[0], colors='blue') 
ax.set_xlim([x_min, x_max]) 
ax.set_ylim([y_min, y_max]) 
ax.set_xlabel('x') 
ax.set_ylabel('y') 
ax.set_title('Графік системи рівнянь') 
plt.grid(True) 
plt.show() 
```

```python
from scipy import optimize
from scipy.misc import derivative
import math

x0 = 0.6
y0 = 0.65
delta = 0.1

def f1(x):
    return 1.5 - math.cos(x)

def f2(y):
    return (math.sin(y - 0.5)) / 2 + 0.5

def iter(x, y, e):
    xn = x
    yn = y
    xn1 = f2(xn)
    yn1 = f1(yn)
    n = 1
    while (abs(xn1 - xn) >= e) and (abs(yn1 - yn) >= e):
        xn = xn1
        yn = yn1
        xn1 = f2(yn)
        yn1 = f1(xn)
        n += 1
    print('Simple iteration:')
    print('x=', xn, '\ny=', yn, '\nThe amount of iteration = ', n)

iter(x0, y0, 0.0001)

def f3(x):
    return math.cos(x[0]) + x[1] - 1.5, 2 * x[0] - math.sin(x[1] - 0.5) - 0.5

s = optimize.root(f3, [0., 0.], method='hybr')
print('Check', s.x)
```
