```python
import numpy as np
from math import factorial 
import matplotlib.pyplot as plt

# задані точки
x_values = np.arange(0.180, 0.231, 0.005)
y_values = np.array([5.5154, 5.4669, 5.3263, 5.1930, 5.0664, 4.9461, 4.8317, 4.7226, 4.6185, 4.5191, 4.4242])

def first_interpolation(x, y, x0):
    n = len(x)
    f = np.zeros((n, n))
    f[:, 0] = y
    for j in range(1, n):
        for i in range(n-j):
            f[i, j] = (f[i+1, j-1]-f[i, j-1])/(x[i+j] - x[i])
    ans = 0
    for j in range(n):
        prod = f[0, j]
        for i in range(j):
            prod *= (x0-x[i])
        ans += prod
    return ans

# друга інтерполяційна формула
def second_interpolation(x, y, x0):
    n = len(x)
    f = np.zeros((n, n))
    f[:, 0] = y
    for j in range(1, n):
        for i in range(n-j):
            f[i, j] = (f[i+1, j-1]-f[i, j-1])/(x[i+j] - x[i])
    ans = f[0, 0]
    for j in range(1, n):
        prod = f[0, j]
        for i in range(j):
            prod *= (x0-x[i])
        ans += prod
    return ans

# Задані x для обчислень
x_values_to_find = [0.184, 0.189, 0.193, 0.221, 0.227, 0.216]

# Знаходження y для заданих x
for x_to_find in x_values_to_find:
    y_to_find = first_interpolation(list(x_values), y_values, x_to_find)
    print(f"f({x_to_find}) = {y_to_find}")
```

```python
xx = np.linspace(np.min(x_values), np.max(x_values), 100)
yy = np.zeros_like(xx)
for i in range(len(xx)):
    yy[i] = second_interpolation(x_values, y_values, xx[i])

plt.plot(x_values, y_values, 'o', label='Дaнi точки')
plt.plot(xx, yy, label='багаточлен Ньютона')
plt.title("Графік інтерполяційної функції Ньютона")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
