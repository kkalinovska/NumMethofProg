```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange
x = np.array([-1., 0., 1., 2.])
y = np.array([-20., -5., 6., 25.])
x_list = [-0.5, 0.5, 1.5, 2.5]
def lagrange_interpolation(x, y, x_test):
    n = len(x)
    p = np.zeros(n)
    for i in range(n):
        p_i = 1
        for j in range(n):
            if i != j:
                p_i *= (x_test - x[j]) / (x[i] - x[j])
        p[i] = p_i
    return np.dot(y, p)
for i in range(len(x_list)):
    x_test = x_list[i]
    f_interp = lagrange_interpolation(x, y, x_test)
    print(f"Значення функції у точці x{i+1}_test = {f_interp:.4f}")
```

```python
xnew=np.linspace(np.min(x),np.max(x),100) 
ynew=[lagrange_interpolation(x,y,i) for i in xnew] 

plt.plot(x, y, 'gs')  
plt.plot(xnew, ynew, 'r')  
plt.title('Lagrange Polynomial_1')
plt.xlabel('Вісь x') 
plt.ylabel('Вісь y')
plt.grid(True)
plt.show()
```

```python
f = lagrange(x, y) 
fig = plt.figure(figsize = (7,5)) 
plt.plot(xnew, f(xnew), 'g', x, y, 'bo') 
plt.title('Lagrange Polynomial_2') 
plt.grid() 
plt.xlabel('Вісь x') 
plt.ylabel('Вісь y') 
plt.show()
```
